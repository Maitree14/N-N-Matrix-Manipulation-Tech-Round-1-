```python
import pandas as pd
```

```python
data=pd.read_csv("analysis_id(in).csv")
data=pd.DataFrame(data)
(data)
```

```python
total_bids=data['bidder'].value_counts().reset_index()
```

```python
total_bids
```

```python
wins_1=data.groupby('bidder')[['request_count','win_count']].sum().reset_index()
```

```python
wins_1
```

```python
wins_1['ratio']=wins_1['win_count']/wins_1['request_count']
```

```python
wins_1
```

```python
final_data_1=wins_1.merge(total_bids,on='bidder',how='left')
final_data_1
```

```python
final_data_1['win_rate_1']=(final_data_1['ratio']/final_data_1['count'])*100
```
